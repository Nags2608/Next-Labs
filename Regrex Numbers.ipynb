```python
#1.1)

# Python3 program to validate
# hexadecimal colour code using
# Regular Expression
import re

# Function to validate
# hexadecimal color code .
def isValidHexaCode(str):

	# Regex to check valid
	# hexadecimal color code.
  regex = "^#([A-Fa-f0-9]{2}|[A-Fa-f0-9]{3})|[A-Fa-f0-9]{1}$"

	# Compile the ReGex
  p = re.compile(regex)

	# If the string is empty
	# return false
  if(str == None):
		 return False 

	# Return if the string
	# matched the ReGex
  if(re.search(p, str)):
	 	return True
  else:
	 	return False

# Driver Code.

# Test Case 1:
str1 = "#1";
print(str1, ":", isValidHexaCode(str1))

# Test Case 2:
str2 = "#2";
print(str2, ":", isValidHexaCode(str2))

# Test Case 3:
str3 = "#3";
print(str3, ":", isValidHexaCode(str3))

# Test Case 4:
str4 = "#4";
print(str4, ":", isValidHexaCode(str4))

# Test Case 5:
str5 = "#5";
print(str5, ":", isValidHexaCode(str5))

# Test Case 6:
str6 = "#6";
print(str6, ":", isValidHexaCode(str6))

# Test Case 7:
str7 = "#7";
print(str7, ":", isValidHexaCode(str7))

# Test Case 8:
str8 = "#8";
print(str8, ":", isValidHexaCode(str8))

# Test Case 9:
str9 = "#9";
print(str9, ":", isValidHexaCode(str9))

# Test Case 10:
str10 = "#10";
print(str10, ":", isValidHexaCode(str10))

# Test Case 11:
str11 = "#11";
print(str11, ":", isValidHexaCode(str11))

# Test Case 12:
str12 = "#648";
print(str12, ":", isValidHexaCode(str12))

# Test Case 13:
str13 = "#649";
print(str13, ":", isValidHexaCode(str13))

# Test Case 14:
str14 = "#650";
print(str14, ":", isValidHexaCode(str14))

# Test Case 15:
str15 = "#651";
print(str15, ":", isValidHexaCode(str15))

# Test Case 16:
str16 = "#652";
print(str16, ":", isValidHexaCode(str16))

# Test Case 17:
str17 = "#653";
print(str17, ":", isValidHexaCode(str17))
```
